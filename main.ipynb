```python
fig, axs = plt.subplots(ncols=2, nrows=3, figsize=(12, 15))

sns.histplot(data=df['Year'], ax=axs[0,0])
axs[0,0].set_title('Distribution of Years Introduced')
axs[0,0].set_ylabel('')

sns.histplot(data=df['speed_mph'], ax=axs[0,1])
axs[0,1].set_xlabel('Speed (mph)')
axs[0,1].set_ylabel('')
axs[0,1].set_title('Distribution of Speed of Rollercoaster (mph)')

sns.kdeplot(data=df['height_ft'], ax=axs[1,0])
axs[1,0].set_title('KDE of height (ft)')
axs[1,0].set_xlabel('Height (ft)')
axs[1,0].set_ylabel('')

sns.histplot(data=df['Gforce'], ax=axs[1,1])
axs[1,0].set_title('Distribution of G Force')
axs[1,0].set_xlabel('G Force')
axs[1,0].set_ylabel('')

sns.histplot(data=df['Inversions'], ax=axs[2,0])
axs[2,0].set_title('Distribution of Inversions')
axs[2,0].set_xlabel('Quantity of Inversions')
axs[2,0].set_ylabel('')

sns.barplot(data=df[['Type', 'Year']].groupby('Type').count().reset_index(), x='Type', y='Year')
axs[2,1].set_title('Distibution of Types of Rollercoasters')
axs[2,1].set_ylabel('')

plt.tight_layout()
plt.show()
```

<h1>Question</h1>
<h3>Which locations feature the fastest roller coasters and offer more than 10 different ride options?</h3>

```python
fig, ax = plt.subplots(ncols=1, nrows=2, figsize=(12,10))

sns.barplot(data= df.query('Location != "Other"').groupby('Location')['speed_mph'] \
            .agg(['mean', 'count']).query('count >= 10').sort_values('mean', ascending=False),
            x= 'mean', y='Location', orient='h', ax=ax[0]
            )
ax[0].set_title('Mean Speed of Coasters by Location')
ax[0].set_xlabel('Mean Speed')
ax[0].set_ylabel('')

sns.barplot(data= df.query('Location != "Other"').groupby('Location')['speed_mph'] \
            .agg(['mean', 'count']).query('count >= 10').sort_values('count', ascending=False),
            x= 'count', y='Location', orient='h', ax=ax[1]
            )
ax[1].set_title('Count of Coasters by Location')
ax[1].set_xlabel('Number of Coasters')
ax[1].set_ylabel('')

plt.show()
```

<h4>
Busch Gardens Williamsburg offers 11 roller coaster options and has the highest average speed overall. However, Cedar Point provides 18 options with a comparable average speed. Therefore, I would choose Cedar Point.
